"""
Two Sum Algorithm

This algorithm finds two numbers in an array that add up to a specific target.
It returns the indices of the two numbers.

Brute Force Approach:
- Time Complexity: O(n^2) - We check each pair of numbers.
- Space Complexity: O(1) - No additional space is used.

Step1: Iterate through each element in the array.
Step2: For each element, iterate through the rest of the array to find the pair sum that matches the target.
Step3: If the pair is found then return the indices of two numbers that add up to target.
Step4: If the pair is not found then iterate through the next number from array and go to Step 2.
"""


```python
def two_sum(nums, target):
    """
    Two Sum Program

    Problem: Given an array of integers nums and an integer target,
    return the indices of the two numbers that add up to the target.

    Args:
        nums (list[int]): Array of integers
        target (int): Target sum

    Returns:
        list[int]: Indices of two numbers that add up to target
    """
    nums_length = len(nums)
    for ind1 in range(nums_length):
        for ind2 in range(ind1 + 1, nums_length):
            if nums[ind1] + nums[ind2] == target:
                return [ind1, ind2]
    return []


if __name__ == "__main__":
    # Test cases
    print(two_sum([2, 7, 11, 15], 9))  # [0, 1]
    print(two_sum([3, 2, 4], 6))       # [1, 2]
    print(two_sum([3, 3], 6))          # [0, 1]
```

Optimised Version of Brute force to use the single iterator but update the initial position based on the last reach of the iterator index when we don't see the target matching with the nums[initial] + nums[ind + 1].

- Time Complexity: O(n^2) - We check each pair of numbers.
- Space Complexity: O(1) - No additional space is used.

```python
def two_sum(nums, target):
    """
    Two Sum Program

    Problem: Given an array of integers nums and an integer target,
    return the indices of the two numbers that add up to the target.

    Args:
        nums (list[int]): Array of integers
        target (int): Target sum

    Returns:
        list[int]: Indices of two numbers that add up to target
    """
    nums_length = len(nums)
    initial = 0
    ind1 = 0
    while ind1 < nums_length - 1:
        if (nums[initial] + nums[ind1 + 1] == target):
            return [initial, ind1 + 1]
        if (ind1 == nums_length - 2):
            initial += 1
            ind1 = initial - 1
        else:
            ind1 += 1
    return []


if __name__ == "__main__":
    # Test cases
    print(two_sum([2, 7, 11, 15], 9))  # [0, 1]
    print(two_sum([3, 2, 4], 6))       # [1, 2]
    print(two_sum([3, 3], 6))          # [0, 1]
```

```python
def two_sum(nums, target):
    """
    Two Sum Program

    Problem: Given an array of integers nums and an integer target,
    return the indices of the two numbers that add up to the target.

    Args:
        nums (list[int]): Array of integers
        target (int): Target sum

    Returns:
        list[int]: Indices of two numbers that add up to target
    """
    nums_length = len(nums)
    resultMap = {}
    results = []
    for ind1, num in enumerate(nums):
        diff = target - num
        if diff in resultMap:
            for ind2 in resultMap[diff]:
                results.append([ind2, ind1])
        if num not in resultMap:
            resultMap[num] = []
        resultMap[num].append(ind1)
    return [item for sublist in results for item in sublist]


if __name__ == "__main__":
    # Test cases
    print(two_sum([2, 7, 11, 15], 9))  # [0, 1]
    print(two_sum([3, 2, 4], 6))       # [1, 2]
    print(two_sum([3, 3], 6))          # [0, 1]
```

Two Pointer approach:

Step 1: Sort the array (nlog(n))
Step 2: initialize the left pointer at start and right pointer at the length - 1
Step 3: store the array enumerate in the index

```python
def two_sum(nums, target):
    """
    Two Sum Program

    Problem: Given an array of integers nums and an integer target,
    return the indices of the two numbers that add up to the target.

    Args:
        nums (list[int]): Array of integers
        target (int): Target sum

    Returns:
        list[int]: Indices of two numbers that add up to target
    """
    left = 0
    right = len(nums) - 1
    indexMap = sorted((num, i) for i, num in enumerate(nums))

    while left < right:
        current_sum = indexMap[left][0] + indexMap[right][0]
        if current_sum == target:
            return [indexMap[left][1], indexMap[right][1]]
        elif current_sum < target:
            left += 1
        else:
            right -= 1

    return []



if __name__ == "__main__":
    # Test cases
    print(two_sum([2, 7, 11, 15], 9))  # [0, 1]
    print(two_sum([3, 2, 4], 6))       # [1, 2]
    print(two_sum([3, 3], 6))          # [0, 1]
```

```python
def two_sum(nums, target):
    """
    Two Sum Program

    Problem: Given an array of integers nums and an integer target,
    return the indices of the two numbers that add up to the target.

    Args:
        nums (list[int]): Array of integers
        target (int): Target sum

    Returns:
        list[int]: Indices of two numbers that add up to target
    """
    obj = {}

    for i in range(len(nums)):
        obj[nums[i]] = i

    for i in range(len(nums)):
        complement = target - nums[i]
        res = obj.get(complement)
        if res is not None and res != i:
            return [i, res]
        
    return []



if __name__ == "__main__":
    # Test cases
    print(two_sum([2, 7, 11, 15], 9))  # [0, 1]
    print(two_sum([3, 2, 4], 6))       # [1, 2]
    print(two_sum([3, 3], 6))          # [0, 1]
```
